## Análise Exploratória de Dados (EDA)

### 1. Visão Geral dos Dados

```python
import pandas as pd
from google.colab import files
import io

print("Por favor, selecione o primeiro arquivo CSV para df1:")
uploaded_df1 = files.upload()

# Assuming only one file is uploaded
for filename in uploaded_df1.keys():
  df1 = pd.read_csv(io.BytesIO(uploaded_df1[filename]))
  print(f"O arquivo '{filename}' foi carregado com sucesso em df1.")
  display(df1.head())
  break # Process only the first uploaded file
```

```python
import pandas as pd
from google.colab import files
import io

print("Por favor, selecione o segundo arquivo CSV para df2:")
uploaded_df2 = files.upload()

# Assuming only one file is uploaded
for filename in uploaded_df2.keys():
  df2 = pd.read_csv(io.BytesIO(uploaded_df2[filename]))
  print(f"O arquivo '{filename}' foi carregado com sucesso em df2.")
  display(df2.head())
  break # Process only the first uploaded file
```

```python
print("--- Informações do DataFrame 1 (df1) ---")
df1.info()
print("\n--- Primeiras 5 linhas do DataFrame 1 (df1) ---")
display(df1.head())
print("\n--- Estatísticas Descritivas do DataFrame 1 (df1) ---")
display(df1.describe())
print("\n--- Contagem de valores nulos por coluna no DataFrame 1 (df1) ---")
display(df1.isnull().sum())
```

## 3. Consistência de Colunas: `tipo_parto` vs `tipo_parto_1`

```python
print("--- df1: Valores únicos e contagem para 'tipo_parto' ---")
display(df1['tipo_parto'].value_counts(dropna=False))

print("\n--- df1: Valores únicos e contagem para 'tipo_parto_1' ---")
display(df1['tipo_parto_1'].value_counts(dropna=False))

print("\n--- df2: Valores únicos e contagem para 'tipo_parto' ---")
display(df2['tipo_parto'].value_counts(dropna=False))

print("\n--- df2: Valores únicos e contagem para 'tipo_parto_1' ---")
display(df2['tipo_parto_1'].value_counts(dropna=False))
```

```python
print("--- Verificando se 'tipo_parto' e 'tipo_parto_1' são idênticos em df1 ---")
display(df1['tipo_parto'].equals(df1['tipo_parto_1']))

print("\n--- Verificando diferenças entre 'tipo_parto' e 'tipo_parto_1' em df1 ---")
display(df1[df1['tipo_parto'] != df1['tipo_parto_1']][['tipo_parto', 'tipo_parto_1']].drop_duplicates())

print("\n--- Verificando se 'tipo_parto' e 'tipo_parto_1' são idênticos em df2 ---")
display(df2['tipo_parto'].equals(df2['tipo_parto_1']))

print("\n--- Verificando diferenças entre 'tipo_parto' e 'tipo_parto_1' em df2 ---")
display(df2[df2['tipo_parto'] != df2['tipo_parto_1']][['tipo_parto', 'tipo_parto_1']].drop_duplicates())
```
