```python
#1. Configuração do Projeto
project_id = "projeto-aplicado-2"

#2. Definição da Query
#Filtrando po Estado (SP) e Ano (2023)
Query = """
SELECT
  ano,
  sigla_uf,
  tipo_parto,
  id_municipio_mae,
  quantidade_filhos_vivos,
  quantidade_parto_normal,
  quantidade_parto_cesareo,
  idade_mae,
  escolaridade_mae,
  raca_cor_mae,
  semana_gestacao,
  data_nascimento,
  tipo_parto
FROM
`basedosdados.br_ms_sinasc.microdados`
WHERE ano = 2023
"""
print("Iniciando a extração dos dados da nuvem... Aguarde.")

# Explicitly authenticate for Colab
from google.colab import auth
auth.authenticate_user()

#3. Execução da busca
df = bd.read_sql(Query, billing_project_id=project_id)

#4. Salvando em CSV
#Oparâmetro index=False evita que o pandas crie uma coluna extra de números.
nome_arquivo = "dados_sinasc_trabalho_projeto.csv"
df.to_csv(nome_arquivo, index=False, encoding='utf-8' )

print(f"Sucesso! O arquivo '{nome_arquivo}' foi gerado com {len(df)} linhas.")
```

```python
import pandas_gbq

project_id = "projeto-aplicado-2"
dataset_id = 'br_ms_sinasc'
table_id_short = 'microdados'

# Query INFORMATION_SCHEMA para puxar as colunas (nome e tipo)
schema_query = f"""
SELECT
  column_name,
  data_type
FROM
  `basedosdados.{dataset_id}.INFORMATION_SCHEMA.COLUMNS`
WHERE
  table_name = '{table_id_short}'
ORDER BY
  ordinal_position
"""

schema_df = pandas_gbq.read_gbq(schema_query, project_id=project_id)

# Display column names and their data types
print('Available columns in the table:')
for index, row in schema_df.iterrows():
    print(f"- {row['column_name']} ({row['data_type']})")
```
